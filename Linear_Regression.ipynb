```python
import pandas as pd   #it is for read, manipulate, and analyze tabular data
import matplotlib.pyplot as plt # Visualization library to create charts and graphs
import seaborn as sns #Built on matplotlib, provides better-looking statistical plots
import numpy as np # Handle karta hai numerical operations
```

```python
from sklearn.linear_model import LinearRegression  #Linear Regression model
```

```python
df = pd.read_csv("home.csv") # readiong the file
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)  #replace infinite values with NaN
df["bedrooms"].fillna(df["bedrooms"].median(), inplace=True)  #fill NaN values with median value of bedrooms
df["bedrooms"] = df["bedrooms"].astype("int64")  #convert bedrooms to integer
```

```python
sns.pairplot(df, diag_kind='kde')   #pairplot for the dataset
plt.show() #
```

```python
X = df.drop(["price"], axis=1) #
y = df["price"]
```

```python
mdl = LinearRegression()
mdl.fit(X, y)
```

```python
y_pred = mdl.predict(X)
```

```python
print("Intercept:", mdl.intercept_)
print("Coefficients:", mdl.coef_)
print("R-squared score:", mdl.score(X, y))
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(y, y_pred, color="green", alpha=0.5)
plt.plot([min(y), max(y)], [min(y), max(y)], linestyle="dashed", color="black")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")

plt.show()
```
