```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/IRIS.csv.xls')
```

```python
df.head()
```

```python
df.shape
```

```python
print("\nBasic Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df.iloc[:, -1])  # Assuming the last column is the target class
plt.title("Class Distribution")
plt.show()
```

```python
# Pairplot to visualize relationships between features
sns.pairplot(df, hue=df.columns[-1])  # Color by the target class
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Boxplots to check for outliers
plt.figure(figsize=(12, 6))
df.iloc[:, :-1].boxplot()
plt.title("Boxplot of Features")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.iloc[:, :-1].boxplot()
plt.title("Boxplot of Features")
plt.show()
```

```python
# Histogram for feature distributions
df.iloc[:, :-1].hist(figsize=(12, 8), bins=15)
plt.suptitle("Feature Distributions")
plt.show()
```
